```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme()

```

```python
rainbow_trout_datapoints = {
    'days': [0, 30, 120], 
    'fasted_weight': [286, 232, 219],
    'fed_weight': [283, 299, 530],
    're-fed_weight': [280, 231, 412],
}
```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(rainbow_trout_datapoints['days'], rainbow_trout_datapoints['fasted_weight'], label='Fasted')
ax.plot(rainbow_trout_datapoints['days'], rainbow_trout_datapoints['fed_weight'], label='Fed')
ax.plot(rainbow_trout_datapoints['days'], rainbow_trout_datapoints['re-fed_weight'], label='Re-fed')

# Add labels
ax.set_xlabel('Days')
ax.set_ylabel('Weight (g)')

# Add legend
ax.legend()

plt.tight_layout()
plt.savefig('rainbow_trout_weight.pdf')

# Show the plot
plt.show()
```

```python
# Compute simple linear coefficients
coefficients = {key: (value[-1] - value[0]) / 120 for key, value in rainbow_trout_datapoints.items() if key != 'days'}

# Print the coefficients per day
print(coefficients)
```

```python
energy_consumption_ps = coefficients['fasted_weight'] / 24 / 60 / 60

# Print the energy consumption per second
print(f'The energy consumption per second is {energy_consumption_ps} g/s')


# Add the growth of the fed fish, together with the energy consumption of the fasted fish
food_consumed = coefficients['fed_weight'] + abs(coefficients['fasted_weight'])
food_consumed_ps = food_consumed / 24 / 60 / 60

# Print the food consumption per second
print(f'The food consumption per second is {food_consumed_ps} g/s')
```
